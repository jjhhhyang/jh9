(a) Classification Decision Tree: Problems and Applications

A Classification Decision Tree solves problems where the goal is to classify items into categories based on their attributes. It uses a tree-like model of decisions where each node represents a feature test, each branch represents an outcome, and each leaf node represents a class prediction.

Real-world applications:

Healthcare: Diagnosing diseases based on patient symptoms and test results.
Marketing: Predicting if a customer will buy a product based on demographic and behavioral data.
Fraud Detection: Identifying fraudulent transactions based on patterns in financial data.
(b) Differences in Predictions: Decision Tree vs. Linear Regression

Classification Decision Tree:

Makes predictions by following a series of yes/no questions (nodes) based on features.
Each path ends at a leaf node, which contains the predicted class (e.g., Hardcover vs. Paperback in the diagram).
Multiple Linear Regression:

Makes predictions by calculating a weighted sum of the input features.
Outputs a continuous numerical value, which can sometimes be rounded or thresholded for classification tasks but isn’t inherently designed for categorization.

Accuracy
Definition: Proportion of true results (both true positives and true negatives) out of all predictions.
Best for: Balanced datasets where false positives and false negatives carry similar consequences.
Example: Weather forecasting—predicting if it will rain. Here, both incorrect "rain" and "no rain" predictions are equally undesirable.

Sensitivity (Recall)
Definition: Proportion of actual positives correctly identified.
Best for: Situations where missing a positive case (false negative) is critical.
Example: Disease diagnosis—detecting cancer. Missing a cancer case could lead to severe health risks.

Specificity
Definition: Proportion of actual negatives correctly identified.
Best for: Situations where false positives (wrongly identifying something as positive) are costly.
Example: Fraud detection—identifying fraudulent credit card transactions. A false positive might annoy customers by flagging legitimate purchases.

Precision
Definition: Proportion of positive predictions that are actually correct.
Best for: Scenarios where false positives are more problematic than false negatives.
Example: Email spam filters—marking legitimate emails as spam (false positive) is worse than missing a spam email (false negative).

```python
import pandas as pd
import numpy as np

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Step 1: Remove the specified columns
ab_reduced = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Step 2: Drop rows with NaN entries
ab_reduced_noNaN = ab_reduced.dropna()

# Step 3: Convert 'Pub year' and 'NumPages' to integers using .loc
ab_reduced_noNaN.loc[:, 'Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN.loc[:, 'NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)

# Step 4: Convert 'Hard_or_Paper' to category type using .loc
ab_reduced_noNaN.loc[:, 'Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Display the first few rows of the processed data
ab_reduced_noNaN.head()
```

##### Step 1: Create an 80/20 split of the dataset


```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (80% training, 20% testing)
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=42
)

# Check the number of observations in each set
print(f"Training set observations: {ab_reduced_noNaN_train.shape[0]}")
print(f"Testing set observations: {ab_reduced_noNaN_test.shape[0]}")
```

##### Step 2: Fit a DecisionTreeClassifier model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Define the target (y) and features (X)
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Target variable: Hard or Paper
X = ab_reduced_noNaN_train[['List Price']]  # Feature: List Price

# Initialize the decision tree classifier with a max depth of 2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Train the decision tree model
clf.fit(X, y)

# Visualize the decision tree using plot_tree
tree.plot_tree(clf, filled=True, feature_names=['List Price'], class_names=['Paper', 'Hard'], rounded=True)
```

##### Step 3: Visualize the Decision Tree
Using tree.plot_tree(clf), you will visualize the tree's decisions based on List Price. The tree will split the data based on the threshold List Price, making predictions about whether the book is hard cover or paperback.

Step 3: How Predictions Are Made
Once the classifier is trained, it makes predictions by following the decision rules formed at each node of the tree. Here's a general breakdown of how predictions are made for the clf2 model:

Root Node: The root node splits the data based on one of the features. For example, the first decision might be based on NumPages or Thick, depending on which feature results in the best split (i.e., the highest information gain).

Subsequent Nodes: The data is further split at each level of the tree based on the other features. Each internal node of the tree compares the feature values against a threshold to decide whether to send the observation down the left or right branch.

Leaf Nodes: When an observation reaches a leaf node, the decision tree classifier assigns a class label (either Hard or Paper). The class is determined by the majority class in that leaf node.

#### 6. Use previously created ab_reduced_noNaN_test to create confusion matrices for clf and clf2. Report the sensitivity, specificity and accuracy for each of the models

Step 1: Make Predictions on Test Data
We will use ab_reduced_noNaN_test to make predictions using both classifiers (clf and clf2).

Step 2: Compute Confusion Matrices
Using sklearn.metrics.confusion_matrix, we will compute the confusion matrices for both models. We will then calculate the sensitivity, specificity, and accuracy from the confusion matrix.

Step 3: Calculate Metrics
Sensitivity (Recall): The ability of the model to correctly identify positive instances (i.e., books that are hardcover).
Specificity: The ability of the model to correctly identify negative instances (i.e., books that are paperback).
Accuracy: The proportion of correct predictions made by the model.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score

# Define the target variable for the test set
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # Target: Hard or Paper

# Make predictions using clf (first model)
y_pred_clf = clf.predict(ab_reduced_noNaN_test[['List Price']])

# Make predictions using clf2 (second model)
y_pred_clf2 = clf2.predict(ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']])

# Compute confusion matrices for both models
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Function to calculate sensitivity, specificity, and accuracy from confusion matrix
def calculate_metrics(cm):
    # True Positives (TP), True Negatives (TN), False Positives (FP), False Negatives (FN)
    TP = cm[1, 1]  # Hardcover, predicted as hardcover
    TN = cm[0, 0]  # Paperback, predicted as paperback
    FP = cm[0, 1]  # Paperback, predicted as hardcover
    FN = cm[1, 0]  # Hardcover, predicted as paperback
    
    # Sensitivity (Recall) = TP / (TP + FN)
    sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
    
    # Specificity = TN / (TN + FP)
    specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
    
    # Accuracy = (TP + TN) / (TP + TN + FP + FN)
    accuracy = (TP + TN) / (TP + TN + FP + FN) if (TP + TN + FP + FN) != 0 else 0
    
    return round(sensitivity, 3), round(specificity, 3), round(accuracy, 3)

# Calculate metrics for both classifiers
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)

# Print the results
print("For clf (List Price only):")
print(f"Sensitivity: {sensitivity_clf}")
print(f"Specificity: {specificity_clf}")
print(f"Accuracy: {accuracy_clf}")

print("\nFor clf2 (NumPages, Thick, List Price):")
print(f"Sensitivity: {sensitivity_clf2}")
print(f"Specificity: {specificity_clf2}")
print(f"Accuracy: {accuracy_clf2}")
```

Metrics Calculation
Sensitivity (Recall): Measures how well the model identifies positive cases. It is calculated as:
Sensitivity = TP / (TP + FN)
Where TP is True Positives and FN is False Negatives.

Specificity: Measures how well the model identifies negative cases. It is calculated as:
Specificity = TN / (TN + FP)
Where TN is True Negatives and FP is False Positives.

Accuracy: Measures the overall correctness of the model. It is calculated as:
Accuracy = (TP + TN) / (TP + TN + FP + FN)
Where TP is True Positives, TN is True Negatives, FP is False Positives, and FN is False Negatives.

To visualize feature importances in a scikit-learn classification decision tree, you can use the .feature_importances_ attribute of the fitted model. This attribute provides a numerical value for each feature, indicating its relative importance in making predictions. The higher the value, the more important that feature is for the model.

For clf2, which was trained using NumPages, Thick, and List Price, you can visualize and report the most important predictor variable using the following code:

```python
import matplotlib.pyplot as plt

# Get the feature importances and feature names from clf2
feature_importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

# Create a bar plot to visualize the feature importances
plt.barh(feature_names, feature_importances)
plt.xlabel('Feature Importance')
plt.title('Feature Importances for clf2')
plt.show()

# Report the most important feature
most_important_feature = feature_names[feature_importances.argmax()]
print(f"The most important feature for clf2 is: {most_important_feature}")
```

#### 9. Describe the differences of interpreting coefficients in linear model regression versus feature importances in decision trees in two to three sentences
In linear regression, the coefficients represent the change in the predicted outcome for each one-unit change in the corresponding predictor variable, holding all other variables constant. These coefficients are directly interpretable as the strength and direction of the relationship between predictors and the target variable. In contrast, feature importances in decision trees indicate how much each feature contributes to reducing uncertainty (e.g., Gini impurity or entropy) in making predictions, but they do not provide a direct relationship between feature values and the target outcome. Instead, they show which features are most influential in splitting the data at each decision node in the tree.
